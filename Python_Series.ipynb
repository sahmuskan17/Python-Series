#Number Guessing Game

```python
import random
correct_number=random.randint(1,20)
print("Welcome to the  number guessing game")
print("I am thinking a number between 1 to 20. You have 5 attemps to win the game. If you win the game you will get 10 points")
attempts=5
while attempts>0:
  num=int(input("Enter the number:"))
  if num==correct_number:
    print("Congratulations! You got 10 points")
    print(f"The correct number is {correct_number}")
    break
  elif num>correct_number:
    print("Your guess number is too high")
  else:
    print("Your number is too low")

  attempts-=1

  if attempts==0:
    print("Game Over")
    print(f"The correct number is {correct_number}")
```

#Mini Project

Create a Simple Calculator that can perform basic operations like addition, substraction, multiplication,division. The user will enter the two numbers. Baesd on the input,call different functions to perform the operation and return the result

```python
def add(num1, num2):
    return num1 + num2

def sub(num1, num2):
    return num1 - num2

def mul(num1, num2):
    return num1 * num2

def div(num1, num2):
    if num2 == 0:  # Handle division by zero
        return "Error! Division by zero."
    return num1 / num2

def calculator(num1, num2, operation):
    if operation == "add":
        result = add(num1, num2)
    elif operation == "sub":
        result = sub(num1, num2)
    elif operation == "mul":
        result = mul(num1, num2)
    elif operation == "div":
        result = div(num1, num2)
    else:
        return "Choose the correct operation (add, sub, mul, div)"

    return f"{num1} {operation} {num2} = {result}"

if __name__ == "__main__":
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    operation = input("Choose operation (add, sub, mul, div): ").strip()

    result = calculator(num1, num2, operation)
    print(result)
```

# Error and Exception Handling in Python

Try And Except Blocks

The try block lets you test a block of code for errors

The except block lets you hanlde the error

The else block lets you execute code when there is no error

The finally block lets you execute code, regardless of the result of then try and except blocks.
*   FileNotFoundError-Raise when trying to open a file that doesn't exist.
*   NameError-Raised when a varaible or function name is not found in the local or global scope
*   ValueError-Raised when a function gets an argument of the right type but inappropiate value
*   ZeroDivisionError-Raised when attempting to divide by zero
*   IOError-Raised when an I/O operation(like opening a file) fails for an I/O related reason.













```python
try:
  x=10/0
except ZeroDivisionError:
  print("You can't divide by zero")
else:
  print(x)
finally:
  print("This is always executed")
```

```python
try:
  with open('non_existent_file.txt','r') as file:
    content=file.read()
except FileNotFoundError:
  print("File not found")
else:
  print(content)
```

```python
a=20
y=5
try:
  a/0
except ZeroDivisionError:
  print("You can't divide by zero")
else:
  print(a/y)
finally:
  print(a/y)
```
